## 02. 데이터 전처리

```python
# 태그제거
print(train['review'][0][:700])
```

```python
examplel = BeautifulSoup(train['review'][0],"html5lib")

examplel.get_text()[:700] 
# get_text() : 태그를 제거하고 문자만 추출
```

```python
#특수문자 공백으로 치환
letters_only = re.sub('[^a-zA-Z]',' ',examplel.get_text()) #문자 제외하고 전부 공백으로 치환
letters_only
```

```python
#전부 소문자로 변경
lower_case = letters_only.lower()
lower_case
```

```python
#공백기준으로 단어 나눔
words = lower_case.split()
words
#from nltk.tokenize import TreebankWordTokenizer
#words = word_tokenize(lowercase)
```

```python
#불용어 처리
nltk.download('stopwords')
```

```python
words = [w for w in words if not w in stopwords.words('english')]
words
```

```python
#stemming(어간추출)
stemmer = PorterStemmer()
words2 = [stemmer.stem(w) for w in words]
print(words2)
```

```python
new_df = []
for i in range(0,10):
  #1. HTML 제거
  review_text = BeautifulSoup(train['review'][i],"html5lib").get_text()
  #2. 특수문자 제거
  letters_only = re.sub('[^a-zA-Z]',' ',review_text)
  #3. 소문자 변환
  words = lower_case.lower().split()
  #4. 불용어 처리
  stops = set(stopwords.words('english'))
  meaningful_words = [w for w in words if not w in stops]
  #6. 어간추출
  stemmer = PorterStemmer()
  stemming_words = [stemmer.stem(w) for w in meaningful_words]
  #7. 문자열 결합
  s = (' '.join(stemming_words))
  new_df.append(s)
```
